#### cicid feature dualplot

```python
import seaborn as sns

def dual_plot(i, col, data1=cici2cici, data2=unsw2cici, label1='cicids', label2='unsw', method=None):
    fig = plt.subplots(figsize=(16, 6), facecolor='white')
    if method != None:
        sns.distplot(data1[col].apply(method), label=label1, hist=False, rug=True)
        sns.distplot(data2[col].apply(method), label=label2, hist=False, rug=True)
    else:
        sns.distplot(data1[col], label=label1, hist=False, rug=True)
        sns.distplot(data2[col], label=label2, hist=False, rug=True)
    
    plt.subplots_adjust(wspace=0.3, hspace=0.3)
    plt.legend()
    plt.savefig(f'./{i}_{col}.png', bbox_inches = 'tight')
```

```python
features = ConcatedCICI.iloc[:,1:78].columns
```

```python
features
```

```python
dual_plot(0,'tot_bwd_pkts')
```

```python
for i,col in enumerate(features):
    dual_plot(i,col)
```
